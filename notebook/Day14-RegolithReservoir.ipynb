```python
paths = lambda filename: [[ eval(l) for l in b.split(' -> ')] 
                                    for b in open(filename, 'r').readlines()]
examples = paths('../data/Day14-pre.txt')
actuals  = paths('../data/Day14.txt')

START_COL, START_ROW = 0,0

def all_points(start, end): 
    (c1,r1),(c2,r2) = start, end
    cs,rs = sorted([c1-500,c2-500]), sorted([r1,r2])
    result = set()
    for c in range(cs[0],cs[1]+1):
        for r in range(rs[0],rs[1]+1):
            result.add((c,r))
    return result

def rocks_from(paths):
    pair = lambda xs: zip(xs, xs[1:])
    return set([point for path in paths for line in pair(path) for point in all_points(*line)])

def drop(continue_check, rocks):
    sc, sr = START_COL,START_ROW
    while(continue_check(sr,rocks)):
        if (sc,sr+1) not in rocks:        # down
            sr += 1                     # fall down
            continue
        if (sc-1,sr+1) not in rocks:      # down left
            sc -= 1
            sr += 1                     # fall down-left
            continue
        if (sc+1,sr+1) not in rocks:      # down right   
            sc += 1
            sr += 1                     # fall down-right
            continue
        rocks.add((sc,sr))              # add sand
        sc,sr = START_COL,START_ROW
        continue
    return rocks

def print_rocks(rocks):
    rows = set([r for c,r in rocks])
    cols = set([c for c,r in rocks])
    result = ""
    for r in range(min(rows), max(rows) + 1): #(min(rows),max(rows)):
        result += f'\n {r}'
        for c in range(min(cols), max(cols) + 1):
            icon = '#' if (c,r) in rocks  else '.'
            result += icon
    print(result)

def part1(paths):
    rocks = rocks_from(paths)
    print_rocks(rocks)
    last_rock_row = max([r for c,r in rocks])
    valid_position = lambda r,_: r <= last_rock_row
    sands = drop(valid_position, rocks)
    print_rocks(sands)
    return len(sands) - len(rocks)

def part2(paths):
    rocks = rocks_from(paths)
    last_rock_row = max([r for c,r in rocks])
    last_rock_col = max([c for c,r in rocks])
    COL_OFFSET = 100
    ROW_OFFSET = 2
    floor = all_points([(-COL_OFFSET, last_rock_row+ROW_OFFSET),(last_rock_col+COL_OFFSET, +ROW_OFFSET)])

    valid_position = lambda _,rocks: (START_COL,START_ROW) not in rocks 
    return drop(valid_position, rocks)


print('Part 1: ',part1(examples))
#for l in rocks_from(examples): print(l)
```
